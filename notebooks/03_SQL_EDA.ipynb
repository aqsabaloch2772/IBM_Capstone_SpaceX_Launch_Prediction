```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("spacex.db")
```

```python
df = pd.read_csv("spacex_cleaned.csv")
df.to_sql("spacex", conn, if_exists="replace", index=False)
```

```python
pd.read_sql_query("SELECT * FROM spacex LIMIT 5;", conn)
```

```python
pd.read_sql_query("SELECT COUNT(*) FROM spacex;", conn)
```

```python
pd.read_sql_query("SELECT COUNT(*) FROM spacex WHERE success = 1;", conn)
```

```python
pd.read_sql_query("""
SELECT core_core, COUNT(*) as flights
FROM spacex
GROUP BY core_core
ORDER BY flights DESC;
""", conn)
```
