This notebook visualizes the **SpaceX Launch Sites** using accurate coordinates.

It includes:
- Loading IBM launch site location dataset  
- Creating a Folium world map  
- Adding launch site markers  
- Adding range circles  
- Adding popups and tooltips  


```python
import pandas as pd
import folium
import math

df = pd.read_csv("launch_site_locations.csv")
df
```

## üåç Create Base Map

```python
launch_map = folium.Map(location=[28.5, -80.6], zoom_start=4)
launch_map
```

## üìå Add Launch Site Markers

```python

for index, row in df.iterrows():
    folium.Marker(
        [row["Latitude"], row["Longitude"]],
        popup=row["Launch Site"],
        tooltip=f"{row['Launch Site']}",
        icon=folium.Icon(color="blue", icon="rocket", prefix='fa')
    ).add_to(launch_map)

launch_map
```

## üîµ Add Circles Around Launch Sites

```python

for index, row in df.iterrows():
    folium.Circle(
        radius=5000,
        location=[row["Latitude"], row["Longitude"]],
        color="red",
        fill=True,
        fill_opacity=0.3,
        popup=f"5 km radius around {row['Launch Site']}"
    ).add_to(launch_map)

launch_map
```
